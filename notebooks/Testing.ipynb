```python
from sklearn.datasets import load_iris
import pandas as pd
import requests

# Load the Iris dataset
iris = load_iris()

# Create a DataFrame
df = (pd.DataFrame(data=iris.data, columns=iris.feature_names)
      .sample(frac=1)
      .reset_index(drop=True))

# Make predictions
url = 'http://127.0.0.1:5000/predict'
for i in range(df.shape[0]):
    data = {'features': df.loc[i].to_list()}
    response = requests.post(url, json=data)
    print(data, response.json())
```

Docker instructions:
1. To create the docker image: docker build -t iris-flask-app .
2. To run the docker container: docker run -p 5000:5000 iris-flask-app
3. To make predictions in the terminal: curl -X POST http://localhost:5000/predict -H "Content-Type: application/json" -d "{\"features\": [5.1, 3.5, 1.4, 0.2]}"
4. In order for the web app to be accessed go to http://localhost:5000/home
5. To use the web app in other devices connected to a common local network, find the corresponding IPv4 Address and use `http:"IP address":5000/home`.
